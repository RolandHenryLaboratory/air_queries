```python
def query_accessions_from_mrn(mrn):
    search_json = {#'name': '',
            'mrn': mrn,
            # 'accNum': '',
            # 'dateRange': {'start':'','end':'','label':''},
            'modality': 'MR',
            'sourceId': 1,
            'profile': 1
    }
    exam_lst = requests.post(urljoin(radURL, 'secure/search/query-data-source'), 
                            headers = header,
                            json = search_json).json()['exams'] 

    accession_set = set()
    for exam_dict in exam_lst:
        if exam_dict['modality'] == 'MR':
            accession_set.add({'accessionNumber' : exam_dict['accessionNumber'],
                               'studyDescription' : exam_dict['studyDescription']})
            
    return accession_set
```
